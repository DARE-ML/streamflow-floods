```python
print(pd.DataFrame(summary_dict).T.to_latex(float_format='%.4f'))
```

```python
import os
import pandas as pd
```

```python
pd.set_option("display.precision", 4)
```

```python
res_dir = 'results/Stage 6/'
```

```python
summary_cols = ['SER_1%', 'SER_2%', 'SER_5%', 'SER_10%', 'SER_25%', 'SER_50%', 'SER_75%', 'RMSE']

summary_list = []


for state_res in os.listdir(res_dir):

    q05_model_res = pd.read_csv(f'{res_dir}/{state_res}/results_q05.csv')
    q95_model_res = pd.read_csv(f'{res_dir}/{state_res}/results_q95.csv')
    reg_model_res = pd.read_csv(f'{res_dir}/{state_res}/results_reg.csv')
    
    reg_model_summary = reg_model_res[summary_cols].mean()
    q05_model_summary = q05_model_res[summary_cols].mean()
    q95_model_summary = q95_model_res[summary_cols].mean()
    
    summary_dict = {'regular': reg_model_summary,
                    'q=0.05': q05_model_summary,
                    'q=0.95': q95_model_summary}

    summary_df = pd.DataFrame(summary_dict).T
    summary_df.index.name = 'Model' 
    summary_df = summary_df.reset_index()

    summary_df.loc[:, 'State'] = state_res.split('-')[0]

    summary_list.append(summary_df)
    

summary_df = pd.concat(summary_list, axis=0)
summary_df = summary_df.set_index(['State', 'Model'])

summary_df
```

```python
print(summary_df.to_latex(float_format='%.4f'))
```
